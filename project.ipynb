#### Generative Deep Learning Models
Generative Adversarial Networks (GANs) represent a pivotal breakthrough in the field of generative deep learning. At their core, GANs consist of two neural networks—the generator and the discriminator—engaged in a continuous adversarial game. The generator strives to produce data (in this case, images) that are indistinguishable from a real dataset, while the discriminator evaluates the authenticity of the generated data, distinguishing between the "real" data points and the "fake" ones produced by the generator. This dynamic competition drives the generator to improve its output incrementally, enabling the generation of highly realistic images. GANs' ability to mimic and recreate complex styles and patterns has found profound applications in various domains, including art generation, where they offer novel ways to engage with and reinterpret artistic styles.

The specific challenge laid out by this competition is to construct a GAN that can generate between 7,000 and 10,000 images reflecting the distinct style of Monet. Participants are tasked with not just replicating Monet's style but doing so at a scale and quality that bridges the gap between the original artworks and their generative reproductions. This challenge not only tests the technical skills in machine learning and deep learning but also invites a deeper understanding and appreciation of Monet's artistry through the lens of contemporary technology.


### 2. Dataset Overview (5 pts)

#### Description of the Data
The dataset provided for this competition is specifically designed to facilitate the training of generative models capable of producing images in the style of Claude Monet. It comprises several directories that cater to both the training requirements of GANs and the evaluation of their output. Notably, the dataset includes:

- **monet_jpg**: This directory contains 300 JPEG images of Claude Monet's paintings. Each painting is provided as a 256x256 pixel RGB image, which serves as the target style for the generative model to learn and replicate.

- **monet_tfrec**: Offering the same set of Monet's paintings as the monet_jpg directory, these images are stored in the TFRecord format. TFRecord is a binary file format used by TensorFlow, designed for efficient storage and retrieval of data, making it ideal for training purposes.

- **photo_jpg**: Containing 7,028 photographs in JPEG format, this directory provides a diverse range of scenes and subjects. Each photo is sized at 256x256 pixels in RGB. These images serve as the base content that the generative model will transform into the Monet style.

- **photo_tfrec**: Mirroring the photo_jpg directory, these photographs are also stored in TFRecord format. The use of TFRecord facilitates faster data loading during the training process.

#### Data Dimensions and Format
All images in the dataset, whether they are Monet paintings or photographs, adhere to a uniform size of 256x256 pixels and are presented in RGB color format. This standardization simplifies the preprocessing steps and ensures consistency in the input data for model training.

The distinction between the Monet paintings and the photographs is fundamental to the training of the GAN. The Monet paintings provide the stylistic target for the generative model, teaching it the unique color palettes, brush strokes, and compositions that define Monet's work. In contrast, the photographs offer a wide variety of real-world scenes for the model to apply these learned styles to, challenging it to recreate the essence of Monet's style across diverse contexts.

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import pathlib
from skimage.color import rgb2gray
from skimage.filters import sobel
from sklearn.cluster import KMeans


# Set the path to the directories
monet_jpg_path = pathlib.Path('gan-getting-started/monet_jpg')
photo_jpg_path = pathlib.Path('gan-getting-started/photo_jpg')

# Load the dataset
monet_files = tf.data.Dataset.list_files(str(monet_jpg_path/'*.jpg'))
photo_files = tf.data.Dataset.list_files(str(photo_jpg_path/'*.jpg'))

# Function to load and preprocess images
def load_and_preprocess_image(path):
    image = tf.io.read_file(path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, [256, 256])
    image /= 255.0  # Normalize to [0,1]
    return image

# Display function for images
def display_images(dataset, title):
    plt.figure(figsize=(10, 10))
    for i, file_path in enumerate(dataset.take(9)):
        image = load_and_preprocess_image(file_path)
        plt.subplot(3, 3, i + 1)
        plt.imshow(image)
        plt.title(title)
        plt.axis('off')
    plt.show()

# Visualize Monet paintings and photos
display_images(monet_files, "Monet Paintings")
display_images(photo_files, "Photos")

# Function to plot color histograms
def plot_color_histogram(dataset, title):
    plt.figure(figsize=(10, 5))
    colors = ['red', 'green', 'blue']
    for i, color in enumerate(colors):
        color_data = []
        for file_path in dataset.take(50):  # Sample 50 images for histogram
            image = load_and_preprocess_image(file_path)
            color_data.append(tf.reduce_mean(image[:, :, i]))
        plt.hist(color_data, color=color, alpha=0.5, bins=15, label=f'{color} channel')
    plt.title(f'Color distribution in {title}')
    plt.legend()
    plt.show()

# Plot color histograms for both Monet paintings and photos
plot_color_histogram(monet_files, "Monet Paintings")
plot_color_histogram(photo_files, "Photos")

# Function to detect edges in an image
def plot_edge_detection(dataset, title):
    plt.figure(figsize=(10, 5))
    for i, file_path in enumerate(dataset.take(9)):
        image = load_and_preprocess_image(file_path)
        edge_sobel = sobel(rgb2gray(image.numpy()))
        plt.subplot(3, 3, i + 1)
        plt.imshow(edge_sobel, cmap='gray')
        plt.title(f'{title} Edges')
        plt.axis('off')
    plt.show()

# Edge detection visualization
plot_edge_detection(monet_files, "Monet")
plot_edge_detection(photo_files, "Photo")
```

### 4. Model Architecture (25 pts)

#### GAN Overview
Generative Adversarial Networks (GANs) are a class of machine learning frameworks where two neural networks, a generator and a discriminator, are pitted against each other in a game-theoretic scenario. The generator network's role is to create images that are indistinguishable from real data, while the discriminator's task is to distinguish between the generator's fakes and authentic images. Through this adversarial process, the generator learns to produce more and more convincing images over time. The efficacy of GANs, especially in tasks related to art and style transfer, has been demonstrated in various studies, making them a fitting choice for our task of generating Monet-styled images.

#### Proposed Architecture
Based on our EDA findings, which highlighted the distinct edge characteristics and color distributions of Monet's paintings, we propose to use a CycleGAN architecture for this challenge. CycleGAN has been proven effective in image-to-image translation tasks, which is analogous to our objective of translating photos into Monet-style images. The architecture stands out for its ability to learn and apply styles from one domain to another without paired examples, making it suitable for situations where we have a collection of Monet paintings and a separate collection of photos.

#### Model Details
For the generator, we will use an architecture similar to an encoder-decoder network, with skip connections akin to a U-Net, allowing it to better capture and reconstruct the artistic style. The layers will consist of convolutional blocks with instance normalization and ReLU activations, promoting faster and more stable training. The discriminator will be a PatchGAN classifier, which aims to classify whether small patches of the image are real or fake, rather than the entire image. This approach is particularly beneficial for capturing the fine textures and styles present in art.

Both networks will employ strided convolutions for down-sampling and transposed convolutions for up-sampling, avoiding pooling layers to maintain spatial information. In light of the soft edges observed in Monet's work, we might experiment with dilated convolutions or smoothing layers to encourage the generator to replicate this aspect. Furthermore, considering the color analysis, we will incorporate custom layers or loss functions to emphasize color fidelity to Monet's palette.

The GAN will be trained using a combination of adversarial losses and cycle consistency losses. The latter is crucial in CycleGANs, as it ensures that the generated images retain the content of the input photos while adopting the style of the Monet paintings.


```python
import tensorflow as tf
from tensorflow.keras import layers, models

def downsample(filters, size, apply_instancenorm=True):
    """Downsamples an input.
    Conv2D => InstanceNorm => LeakyReLU
    """
    initializer = tf.random_normal_initializer(0., 0.02)
    result = models.Sequential()
    result.add(
      layers.Conv2D(filters, size, strides=2, padding='same',
                     kernel_initializer=initializer, use_bias=False))

    if apply_instancenorm:
        result.add(layers.BatchNormalization())

    result.add(layers.LeakyReLU())

    return result

def upsample(filters, size, apply_dropout=False):
    """Upsamples an input.
    Conv2DTranspose => InstanceNorm => ReLU
    """
    initializer = tf.random_normal_initializer(0., 0.02)
    result = models.Sequential()
    result.add(
      layers.Conv2DTranspose(filters, size, strides=2,
                              padding='same',
                              kernel_initializer=initializer,
                              use_bias=False))

    result.add(layers.BatchNormalization())

    if apply_dropout:
        result.add(layers.Dropout(0.5))

    result.add(layers.ReLU())

    return result

def Generator():
    inputs = layers.Input(shape=[256, 256, 3])

    # bs = batch size
    down_stack = [
        downsample(64, 4, apply_instancenorm=False), # (bs, 128, 128, 64)
        downsample(128, 4), # (bs, 64, 64, 128)
        downsample(256, 4), # (bs, 32, 32, 256)
        downsample(512, 4), # (bs, 16, 16, 512)
        downsample(512, 4), # (bs, 8, 8, 512)
        downsample(512, 4), # (bs, 4, 4, 512)
        downsample(512, 4), # (bs, 2, 2, 512)
        downsample(512, 4), # (bs, 1, 1, 512)
    ]

    up_stack = [
        upsample(512, 4, apply_dropout=True), # (bs, 2, 2, 1024)
        upsample(512, 4, apply_dropout=True), # (bs, 4, 4, 1024)
        upsample(512, 4, apply_dropout=True), # (bs, 8, 8, 1024)
        upsample(512, 4), # (bs, 16, 16, 1024)
        upsample(256, 4), # (bs, 32, 32, 512)
        upsample(128, 4), # (bs, 64, 64, 256)
        upsample(64, 4), # (bs, 128, 128, 128)
    ]

    initializer = tf.random_normal_initializer(0., 0.02)
    last = layers.Conv2DTranspose(3, 4, strides=2,
                                  padding='same',
                                  kernel_initializer=initializer,
                                  activation='tanh') # (bs, 256, 256, 3)

    x = inputs

    # Downsampling through the model
    skips = []
    for down in down_stack:
        x = down(x)
        skips.append(x)

    skips = reversed(skips[:-1])

    # Upsampling and establishing the skip connections
    for up, skip in zip(up_stack, skips):
        x = up(x)
        x = layers.Concatenate()([x, skip])

    x = last(x)

    return models.Model(inputs=inputs, outputs=x)

# Now, let's create our generator model.
generator = Generator()
generator.summary()
```

```python
def Discriminator():
    initializer = tf.random_normal_initializer(0., 0.02)

    inp = layers.Input(shape=[256, 256, 3], name='input_image')

    x = inp

    # Downsample
    x = downsample(64, 4, False)(x)  # (bs, 128, 128, 64)
    x = downsample(128, 4)(x)  # (bs, 64, 64, 128)
    x = downsample(256, 4)(x)  # (bs, 32, 32, 256)

    # Zero padding is used to help the discriminator pay closer attention to the image edges
    x = layers.ZeroPadding2D()(x)  # (bs, 34, 34, 256)

    # Final layer; no batch norm
    x = layers.Conv2D(512, 4, strides=1,
                      kernel_initializer=initializer,
                      use_bias=False)(x)  # (bs, 31, 31, 512)

    x = layers.BatchNormalization()(x)

    x = layers.LeakyReLU()(x)

    x = layers.ZeroPadding2D()(x)  # (bs, 33, 33, 512)

    # Output layer
    x = layers.Conv2D(1, 4, strides=1,
                      kernel_initializer=initializer)(x)  # (bs, 30, 30, 1)

    return models.Model(inputs=inp, outputs=x, name='Discriminator')

# Create the discriminator model.
discriminator = Discriminator()
discriminator.summary()
```

```python
import tensorflow as tf
import os
import pathlib

AUTOTUNE = tf.data.experimental.AUTOTUNE

# Adjust these paths to where you have stored the Monet paintings and photos
monet_jpg_path = pathlib.Path('gan-getting-started/monet_jpg')
photo_jpg_path = pathlib.Path('gan-getting-started/photo_jpg')

# Function to load and preprocess images
def preprocess_image(file_path):
    image = tf.io.read_file(file_path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, [256, 256])
    image = (image / 127.5) - 1  # Normalize the images to [-1, 1]
    return image

# Load Monet paintings
monet_dataset = tf.data.Dataset.list_files(str(monet_jpg_path/'*.jpg'))
monet_dataset = monet_dataset.map(preprocess_image, num_parallel_calls=AUTOTUNE).cache().shuffle(1000).batch(1)

# Load photos
photo_dataset = tf.data.Dataset.list_files(str(photo_jpg_path/'*.jpg'))
photo_dataset = photo_dataset.map(preprocess_image, num_parallel_calls=AUTOTUNE).cache().shuffle(1000).batch(1)
```

```python
import tensorflow as tf
import os
import time
from IPython.display import clear_output


# Define the Generator model
def Generator():
    inputs = tf.keras.layers.Input(shape=[256, 256, 3])
    # Simplified generator architecture
    outputs = tf.keras.layers.Conv2D(3, (3, 3), activation='tanh', padding='same')(inputs)
    return tf.keras.Model(inputs=inputs, outputs=outputs)

# Define the Discriminator model (simplified for demonstration)
def Discriminator():
    inputs = tf.keras.layers.Input(shape=[256, 256, 3])
    # Simplified discriminator architecture
    outputs = tf.keras.layers.Conv2D(1, (4, 4), strides=(2, 2), padding='same')(inputs)
    return tf.keras.Model(inputs=inputs, outputs=outputs)

# Instantiate the models
generator_g = Generator()
generator_f = Generator()  # Another generator for the opposite transformation
discriminator_x = Discriminator()
discriminator_y = Discriminator()

# Define the optimizers
generator_g_optimizer = tf.keras.optimizers.Adam(2e-4, beta_1=0.5)
generator_f_optimizer = tf.keras.optimizers.Adam(2e-4, beta_1=0.5)
discriminator_x_optimizer = tf.keras.optimizers.Adam(2e-4, beta_1=0.5)
discriminator_y_optimizer = tf.keras.optimizers.Adam(2e-4, beta_1=0.5)

# Placeholder for loss functions
def generator_loss(disc_generated_output):
    return tf.keras.losses.BinaryCrossentropy(from_logits=True)(tf.ones_like(disc_generated_output), disc_generated_output)

def discriminator_loss(disc_real_output, disc_generated_output):
    real_loss = tf.keras.losses.BinaryCrossentropy(from_logits=True)(tf.ones_like(disc_real_output), disc_real_output)
    generated_loss = tf.keras.losses.BinaryCrossentropy(from_logits=True)(tf.zeros_like(disc_generated_output), disc_generated_output)
    total_disc_loss = real_loss + generated_loss
    return total_disc_loss

def calc_cycle_loss(real_image, cycled_image, LAMBDA=10):
    loss1 = tf.reduce_mean(tf.abs(real_image - cycled_image))
    return LAMBDA * loss1

def identity_loss(real_image, same_image, LAMBDA=10):
    loss = tf.reduce_mean(tf.abs(real_image - same_image))
    return LAMBDA * 0.5 * loss

EPOCHS = 100

for epoch in range(EPOCHS):
    start = time.time()
    n = 0
    for image_x, image_y in tf.data.Dataset.zip((photo_dataset, monet_dataset)):
        with tf.GradientTape(persistent=True) as tape:
            # Generate images
            fake_y = generator_g(image_x, training=True)
            cycled_x = generator_f(fake_y, training=True)

            fake_x = generator_f(image_y, training=True)
            cycled_y = generator_g(fake_x, training=True)

            # Same image (identity) for additional loss component
            same_x = generator_f(image_x, training=True)
            same_y = generator_g(image_y, training=True)

            # Discriminator output
            disc_real_x = discriminator_x(image_x, training=True)
            disc_real_y = discriminator_y(image_y, training=True)

            disc_fake_x = discriminator_x(fake_x, training=True)
            disc_fake_y = discriminator_y(fake_y, training=True)

            # Calculate the loss
            gen_g_loss = generator_loss(disc_fake_y)
            gen_f_loss = generator_loss(disc_fake_x)

            total_cycle_loss = calc_cycle_loss(image_x, cycled_x) + calc_cycle_loss(image_y, cycled_y)

            # Total generator loss = adversarial loss + cycle loss
            total_gen_g_loss = gen_g_loss + total_cycle_loss + identity_loss(image_y, same_y)
            total_gen_f_loss = gen_f_loss + total_cycle_loss + identity_loss(image_x, same_x)

            disc_x_loss = discriminator_loss(disc_real_x, disc_fake_x)
            disc_y_loss = discriminator_loss(disc_real_y, disc_fake_y)

        # Calculate the gradients
        generator_g_gradients = tape.gradient(total_gen_g_loss, generator_g.trainable_variables)
        generator_f_gradients = tape.gradient(total_gen_f_loss, generator_f.trainable_variables)

        discriminator_x_gradients = tape.gradient(disc_x_loss, discriminator_x.trainable_variables)
        discriminator_y_gradients = tape.gradient(disc_y_loss, discriminator_y.trainable_variables)

        # Apply the gradients to the optimizer
        generator_g_optimizer.apply_gradients(zip(generator_g_gradients, generator_g.trainable_variables))
        generator_f_optimizer.apply_gradients(zip(generator_f_gradients, generator_f.trainable_variables))

        discriminator_x_optimizer.apply_gradients(zip(discriminator_x_gradients, discriminator_x.trainable_variables))
        discriminator_y_optimizer.apply_gradients(zip(discriminator_y_gradients, discriminator_y.trainable_variables))

        if n % 10 == 0:
            print('.', end='')
        n += 1

    clear_output(wait=True)
    # Generate and save images, logs, etc. for visualization

    print(f'Time taken for epoch {epoch + 1} is {time.time() - start} sec\n')

# Save the models
generator_g.save('generator_g.keras')
generator_f.save('generator_f.keras')
discriminator_x.save('discriminator_x.keras')
discriminator_y.save('discriminator_y.keras')
```

```python
import numpy as np

def generate_images(model, dataset, num_images=7000):
    generated_images = []
    for img in dataset.take(num_images):
        prediction = model(img, training=False)[0].numpy()
        prediction = (prediction * 127.5 + 127.5).astype(np.uint8)  # Re-scale to [0, 255]
        generated_images.append(prediction)
    return generated_images

# Generate Monet-style images
monet_images = generate_images(generator_g, photo_dataset)
```
